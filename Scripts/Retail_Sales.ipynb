# Load the dataset

```python
import pandas as pd

# Load the dataset
df = pd.read_excel('Retail_Sales.xlsx')

# Display the first few rows
df.head()
```

```python
# Check the data types and missing values
df.info()

# Display summary statistics
df.describe(include='all')
```

```python
# Check for missing values
df.isnull().sum()

# For simplicity, let's drop rows with any missing values
# Alternatively, you can fill missing values using df.fillna()
df_clean = df.dropna()

# Verify that there are no missing values left
df_clean.isnull().sum()
```

##### Sales Distribution:

```python
import matplotlib.pyplot as plt

# Plotting sales distribution
plt.figure(figsize=(10, 6))
plt.hist(df['SALES'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Sales Amounts')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

##### Product Line Popularity

```python
# Plotting product line distribution
plt.figure(figsize=(10, 6))
df['PRODUCTLINE'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title('Product Line Popularity')
plt.xlabel('Product Line')
plt.ylabel('Number of Orders')
plt.show()
```

#####  Order Status Distribution

```python
# Plotting order status distribution
plt.figure(figsize=(10, 6))
df['STATUS'].value_counts().plot(kind='bar', color='green', edgecolor='black')
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.show()
```
